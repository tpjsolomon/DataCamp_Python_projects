```python
#Â Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
crimes = pd.read_csv("crimes.csv", parse_dates=["Date Rptd", "DATE OCC"], dtype={"TIME OCC": str})
crimes.head()
```

## Which hour has the highest frequency of crimes? 
Store as an integer variable called peak_crime_hour.

```python
# Create a new column called HOUR OCC by extracting the hour (the first two digits) from TIME OCC and convert it to an integer
crimes['HOUR OCC'] = crimes['TIME OCC'].str[:2].astype(int)

# View the dataframe to check the column is correct
crimes.head()
```

```python
# Plot the counts to confirm the largest frequency of crimes per hour is correct.
sns.countplot(data=crimes, x="HOUR OCC").set(title='Number of crimes per hour of the day.')
plt.show()
```

```python
# The plot shows that midday (hour = 12) has the greatest number of crimes.
peak_crime_hour = 12
```

## Which area has the largest frequency of night crimes (crimes committed between 10pm and 3:59am)? 
Save as a string variable called peak_night_crime_location.

```python
# Filter the number crimes committed to the hours between 10pm and 3:59am
night_time = crimes[crimes["HOUR OCC"].isin([22,23,0,1,2,3])]

# Group the number of crimes by "AREA NAME" and filter for the largest value. Save as a string variable called peak_night_crime_location.
peak_night_crime_location = night_time.groupby("AREA NAME", as_index=False)["HOUR OCC"].count().sort_values("HOUR OCC", ascending=False).iloc[0]["AREA NAME"]

# Print the result
print(f"The area of Los Angelese with the largest number of crimes at night is {peak_night_crime_location}.")
```

## Identify the number of crimes committed against victims of different age groups. 
Save as a pandas Series called victim_ages, with age group labels "0-17", "18-25", "26-34", "35-44", "45-54", "55-64", and "65+" as the index and the frequency of crimes as the values.

```python
# Identify the number of crimes committed against victims of different age groups, with age group labels "0-17", "18-25", "26-34", "35-44", "45-54", "55-64", and "65+" as the index and the number of crimes as the values.
crimes['Age Bracket'] = pd.cut(crimes['Vict Age'], bins=[0, 17, 25, 34, 44, 54, 64, np.inf], labels=["0-17", "18-25", "26-34", "35-44", "45-54", "55-64", "65+"])
# Save as a pandas Series called victim_ages
victim_ages = crimes['Age Bracket'].value_counts()

# Print the result
print(victim_ages)
```
